```python
b=a[['香调','年份','标签']]
b['香调']=b['香调'].map(lambda x:x.strip().split())
c = b.explode('香调')
d=c.groupby(['年份','香调']).count().reset_index()
d['年份']=d['年份'].apply(lambda x :str(x)+'/01/01')
```

```python
d.groupby('香调').sum()
```

```python
d=d[d['香调'].isin(['绿叶','醛香'])!=True]
```

```python
# 主题河流示例

from pyecharts.charts import ThemeRiver
import pyecharts.options as opts

#数据格式
# 系列名称，用于 tooltip 的显示，legend 的图例筛选。
#    series_name: Sequence,
    # 系列数据项
#    data: types.Sequence[types.Union[opts.ThemeRiverItem, dict]],

data_list = []
# 封装数据
for i in zip(d['年份'],d['标签'],d['香调']):
    data_list.append(list(i))
# print(data_list)
series = d['香调'].drop_duplicates().to_list()
# 绘制，设置类型为时间
wc = ThemeRiver(init_opts=opts.InitOpts(height='600px'))\
    .add(
        series_name=series,
        data=data_list,
        singleaxis_opts=opts.SingleAxisOpts(
            pos_top="50", pos_bottom="50", type_="time"
        ),
    )\
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))\
    .set_global_opts(xaxis_opts = opts.AxisOpts(is_show = False),yaxis_opts = opts.AxisOpts(is_show = False)) 

wc.render_notebook()
```

```python
a['香材']=a['前调']+' '+a['中调']+' '+a['后调']
a['香材']=a['香材'].map(lambda x:x.strip().split())
```

```python
b = a[['种类','香调','香材']].explode('香材')
b['香调']=b['香调'].map(lambda x:x.strip().split())
c = b.explode('香调')
```

```python
d = c.groupby(['种类','香调']).value_counts().reset_index().sort_values(0)
e = d[d[0]>1].sort_values(['种类','香调','香材'])
e.columns = ['种类','香调','香材','典型程度']
```
